# Seção 2 - Concatenando múltiplos datasets

## Pergunta #8 - Como juntar múltiplos daatasets?

### Passo 1 - Carregando todos os dados

```python
nomes_arquivos = glob.glob("dados-world-happiness/*.csv")
nomes_arquivos
```

```python
dfs = {}
```

```python
# Duas formas diferentes para conseguir o ano do arquivo
nomes_arquivos[0].split("/")[-1].split(".")[0]
nomes_arquivos[0][-8:-4]
```

```python
# Armazena os anos em um dicionário chamado "dfs"
for arquivo in nomes_arquivos:
    dados = pd.read_csv(arquivo)
    ano = arquivo[-8:-4]
    dfs[ano] = dados
```

## Pergunta #9 - Existe dados da Região apenas do ano 2015. Existe uma maneira de trazer dados da Região para os anos em que não os temos?

```python
# Separando um dataframe com o nome dos países e região.
mapping_regiao_df = dfs["2015"][["Country or region", "Region"]]
```

```python
mapping_regiao_df
```

```python
# Removendo a coluna região do dataframe principal.
df_principal = df_principal.drop(columns=["Region"])
```

```python
df_principal
```

```python
# Está sendo feito o merge pela coluna "Country or region", utilizando o método de junção "left" (esquerda), o que
# significa que todas as linhas de df_principal serão mantidas e as informações de map_regiao_df serão 
# adicionadas a elas, caso existam correspondências na coluna "Country or region"

df_principal = pd.merge(left = df_principal, right = mapping_regiao_df, how = "left", on = "Country or region")
```

```python
df_principal
```

## Pergunta #16 - Quais países tiveram sua pontuação mais aumentada ao longo do tempo?

```python
df_principal.head()
```

```python
df_pivot = df_principal.pivot_table(values = "Score", index = ["Region", "Country or region"], columns = "Year")
df_pivot
```

```python
df_pivot["Difference"] = df_pivot["2019"] - df_pivot["2015"]
df_pivot.head()
```

```python
df_pivot.sort_values(by = "Difference", ascending = False)
```

##### Benim, Costa do Marfim, Togo, Honduras e Burquina Fasso foram os que países tiveram sua pontuação de felcidade mais aumentada ao longo do tempo.
