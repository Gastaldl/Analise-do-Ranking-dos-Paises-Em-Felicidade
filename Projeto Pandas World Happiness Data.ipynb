# Seção 2 - Concatenando múltiplos datasets

## Pergunta #8 - Como juntar múltiplos daatasets?

### Passo 1 - Carregando todos os dados

```python
nomes_arquivos = glob.glob("dados-world-happiness/*.csv")
nomes_arquivos
```

```python
dfs = {}
```

```python
# Duas formas diferentes para conseguir o ano do arquivo
nomes_arquivos[0].split("/")[-1].split(".")[0]
nomes_arquivos[0][-8:-4]
```

```python
for arquivo in nomes_arquivos:
    dados = pd.read_csv(arquivo)
    ano = arquivo[-8:-4]
    dfs[ano] = dados
```

```python
dfs["2015"]
```

### Passo 2 - Olhando a forma de cada DataFrame de cada ano

```python
for ano, dados in dfs.items():
    print(f"{ano}: {dados.shape}")
```

### Passo 3 - Adcionando uma coluna "Year" para cada DataFrame

```python
df_copia["Year"] = 2019
df_copia.head()
```

```python
dfs.keys()
```

```python
for ano in dfs.keys():
    dfs[ano]["Year"] = ano
```

```python
dfs["2015"].head()
```

### Passo 4 - Juntando os Dataframes de 2019 e de 2018

```python
# Verificando se as colunas são iguais 
set(dfs["2019"].columns) ^ set(dfs["2018"].columns)
set(dfs["2019"].columns) == set(dfs["2018"].columns)
```

```python
df_19_18 = pd.concat([dfs["2019"], dfs["2018"]], ignore_index = True)
```

```python
df_19_18
```

### Passo 5 - Juntando dados dos anos 2019, 2018 e 2015

```python
set(dfs["2019"].columns) ^ set(dfs["2015"].columns)
```

```python
dfs["2019"].head(1)
```

```python
dfs["2015"].head(1)
```

```python
colunas_novas_2015 = [
    "Country or region", "Region", "Overall rank", "Score", "Standard Error",
    "GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life choices",
    "Perceptions of corruption", "Generosity", "Dystopia Residual", "Year"
]
```

```python
dfs["2015"].columns = colunas_novas_2015
```

```python
df_19_18_15 = pd.concat([dfs["2019"], dfs["2018"], dfs["2015"]], ignore_index = True)
df_19_18_15
```

## Pergunta #9 - Existe dados da Região apenas do ano 2015. Existe uma maneira de trazer dados da Região para os anos em que não os temos?

```python
map_regiao_df = dfs["2015"][["Country or region", "Region"]]
```

```python
map_regiao_df
```

```python
df_19_18_15 = df_19_18_15.drop(columns=["Region"])
```

```python
df_19_18_15
```

```python
df_com_region = pd.merge(left = df_19_18_15, right = map_regiao_df, how = "left", on = "Country or region")
```

```python
df_com_region
```

## Pergunta #10 - Existe alguma região que não correspondeu?

```python
bool_region_mask.sum()
```

```python
df_com_region[bool_region_mask]
```
