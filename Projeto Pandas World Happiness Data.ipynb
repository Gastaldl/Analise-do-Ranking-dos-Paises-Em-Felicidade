# Seção 2 - Concatenando múltiplos datasets

## Pergunta #8 - Como juntar múltiplos daatasets?

### Passo 1 - Carregando todos os dados

```python
nomes_arquivos = glob.glob("dados-world-happiness/*.csv")
nomes_arquivos
```

```python
dfs = {}
```

```python
# Duas formas diferentes para conseguir o ano do arquivo
nomes_arquivos[0].split("/")[-1].split(".")[0]
nomes_arquivos[0][-8:-4]
```

```python
for arquivo in nomes_arquivos:
    dados = pd.read_csv(arquivo)
    ano = arquivo[-8:-4]
    dfs[ano] = dados
```

```python
dfs["2015"]
```

### Passo 2 - Olhando a forma de cada DataFrame de cada ano

```python
for ano, dados in dfs.items():
    print(f"{ano}: {dados.shape}")
```

### Passo 3 - Adcionando uma coluna "Year" para cada DataFrame

```python
df_copia["Year"] = 2019
df_copia.head()
```

```python
dfs.keys()
```

```python
for ano in dfs.keys():
    dfs[ano]["Year"] = ano
```

```python
dfs["2015"].head()
```

### Passo 4 - Juntando os Dataframes de 2019 e de 2018

```python
# Verificando se as colunas são iguais 
set(dfs["2019"].columns) ^ set(dfs["2018"].columns)
set(dfs["2019"].columns) == set(dfs["2018"].columns)
```

```python
df_19_18 = pd.concat([dfs["2019"], dfs["2018"]], ignore_index = True)
```

```python
df_19_18
```

### Passo 5 - Juntando dados dos anos 2019, 2018 e 2015

```python
set(dfs["2019"].columns) ^ set(dfs["2015"].columns)
```
